```python
import pandas as pd
import numpy as np
```

```python
fielding_2023 = pd.read_csv("C:/Users/ASUS/OneDrive/Documents/實習/CPBL/原始資料/Fielding/2023.csv")
```

```python
fielding_2023 = fielding_2023[fielding_2023['POS'] != 'P']
```

```python
fielding_2023.rename(columns={'Name': '名稱'}, inplace=True)
fielding_2023.rename(columns={'Team Name': '球隊'}, inplace=True)
```

```python
fielding_2023
```

```python
def calculate_f6(row):
    if row['POS'] == 'P':
        return 0
    elif row['POS'] == 'C':
        return 1
    elif row['POS'] == '1B':
        return 2
    elif row['POS'] == '2B':
        return ((row['PO']+row['A'] / row['G']) / 4.8) * 6
    elif row['POS'] == '3B':
        return ((row['PO']+row['A'] / row['G']) / 2.65) * 4
    elif row['POS'] == 'SS':
        return ((row['PO']+row['A'] / row['G']) / 4.6) * 7
    elif row['POS'] in ['LF', 'CF', 'RF']:
        return ((row['PO']+row['A'] / row['G']) / 2.0) * 6
    else:
        return None

fielding_2023['F6'] = fielding_2023.apply(calculate_f6, axis=1)
```

```python
fielding=fielding_2023.groupby(['名稱','球隊']).agg({'F6':'mean'})
```

```python
excel_file_path = 'C:/Users/ASUS/OneDrive/Documents/實習/Rebas/CPBL_2023.xlsx'  # 請替換成你的 Excel 檔案路徑
sheet_name1 = 'AWAY-BOX-打擊'  # 請替換成你要載入的工作表名稱
sheet_name2 = 'HOME-BOX-打擊'
```

```python
away = pd.read_excel(excel_file_path, sheet_name=sheet_name1)
home = pd.read_excel(excel_file_path, sheet_name=sheet_name2)
```

```python
merged_stat = pd.concat([away, home], ignore_index=True)
merged_stat.info()
```

```python
merged_stat = merged_stat.drop(["球場","比賽編號","比賽日期","客場","主場","打序","背號",'雙殺','三殺','失誤上壘','打席','打點','犧短','犧飛','故意四壞'], axis = 1)
merged_stat
```

```python
agg_dict = {
    "打數": "sum",
    "得分": "sum",
    "安打": "sum",
    "二安": "sum",
    "三安": "sum",
    "全壘打": "sum",
    "滾地雙殺": "sum",
    "四壞": "sum",
    "觸身": "sum",
    "三振": "sum",
    "盜壘": "sum",
    "盜壘失敗": "sum"
}
batter_total = merged_stat.groupby("名稱").agg(agg_dict).reset_index()
batter_total
```

```python
batter_total['一安']=batter_total['安打']-batter_total['二安']-batter_total['三安']-batter_total['全壘打']
```

```python
batter_total['F1']=((batter_total['盜壘']+3)/(batter_total['盜壘']+batter_total['盜壘失敗']+7)-0.4)*20
batter_total['F2']=np.sqrt((batter_total['盜壘']+batter_total['盜壘失敗'])/(batter_total['一安']+batter_total['四壞']+batter_total['觸身']))/0.7
batter_total['F3']=batter_total['三安']/(batter_total['打數']-batter_total['全壘打']-batter_total['三振'])/0.0016
batter_total['F4']=25*(((batter_total['得分']-batter_total['全壘打'])/(batter_total['安打']+batter_total['四壞']+batter_total['觸身']-batter_total['全壘打']))-0.1)
batter_total['F5']=(0.063-(batter_total['滾地雙殺']/(batter_total['打數']-batter_total['全壘打']-batter_total['三振'])))/0.007                      
```

```python
batter_total
```

```python
result= pd.merge(batter_total,fielding, on='名稱', how='outer')
result
```

```python
result[['F1', 'F2', 'F3', 'F4', 'F5', 'F6']] = result[['F1', 'F2', 'F3', 'F4', 'F5', 'F6']].clip(lower=0, upper=10)
result[['F1', 'F2', 'F3', 'F4', 'F5', 'F6']] = result[['F1', 'F2', 'F3', 'F4', 'F5', 'F6']].fillna(0)
```

```python
result
```

```python
result['Speed Score']=(result['F1']+result['F2']+result['F3']+result['F4']+result['F5']+result['F6'])/6
```
